```python
def weighted_average(values):
    # Calculate the weighted sum
    weighted_sum = sum(value * value for value in values)

    # Calculate the sum of all values
    total_value = sum(values)

    # Calculate the weighted average
    if total_value != 0:
        weighted_avg = weighted_sum / total_value
    else:
        weighted_avg = 0  # Handle the case where the sum of values is 0

    return weighted_avg  / total_value * 100
```
